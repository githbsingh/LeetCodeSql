#### Q. 1934. Confirmation Rate
Sol. SELECT s.user_id, coalesce(round(src.confirmation_rate,2),0) as confirmation_rate FROM Signups s LEFT JOIN
(SELECT user_id, avg(confirmation_rate) as confirmation_rate FROM (SELECT user_id,
    CASE WHEN action = 'confirmed' THEN 1
    ELSE 0 END as confirmation_rate
FROM Confirmations) as src
GROUP BY user_id) as src
ON src.user_id=s.user_id
